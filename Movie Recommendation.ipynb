```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep='\t', names=column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv("Movie_Id_Titles")
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles, on='item_id')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
plt.figure(figsize=(10,4))
ratings['num of ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating',y='num of ratings',data=ratings,alpha=0.5)
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
ratings.head()
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
toystory_user_ratings = moviemat['Toy Story (1995)']
pulpfiction_user_ratings = moviemat['Pulp Fiction (1994)']
homealone_user_ratings = moviemat['Home Alone (1990)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_toystory = moviemat.corrwith(toystory_user_ratings)
similar_to_pulpfiction = moviemat.corrwith(pulpfiction_user_ratings)
similar_to_homealone = moviemat.corrwith(homealone_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_toystory = pd.DataFrame(similar_to_toystory,columns=['Correlation'])
corr_toystory.dropna(inplace=True)
corr_toystory = corr_toystory.join(ratings['num of ratings'])
corr_toystory[corr_toystory['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_pulpfiction = pd.DataFrame(similar_to_pulpfiction,columns=['Correlation'])
corr_pulpfiction.dropna(inplace=True)
corr_pulpfiction = corr_pulpfiction.join(ratings['num of ratings'])
corr_pulpfiction[corr_pulpfiction['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_homealone = pd.DataFrame(similar_to_homealone,columns=['Correlation'])
corr_homealone.dropna(inplace=True)
corr_homealone = corr_homealone.join(ratings['num of ratings'])
corr_homealone[corr_homealone['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
